```python
# Creating SQLite Database
from sqlalchemy import create_engine, Column, Integer, String, Float, Text
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import Session
import pandas as pd
import numpy as np
Base = declarative_base()
```

```python
# File to Load
btc_data = "7mos_test_data.csv"
#BitMEX_data = "BitMEX_Funding_Data.csv"

# Read the modified csv and store into Pandas DataFrame
btc_df = pd.read_csv(btc_data, encoding="utf-8")
#BitMEX_df = pd.read_csv(BitMEX_data, encoding="utf-8")
btc_df.head()
```

```python
type(btc_df['rolling_20_d'][1])
```

```python
btc_df['rolling_20_d'] = pd.to_numeric(btc_df['rolling_20_d'])
```

```python
# Define the happiness table
class btc(Base):
    __tablename__ = "btc"
    id =Column(Integer, primary_key=True)
    date = Column(String)
    open = Column(Float)
    high = Column(Float)
    low = Column(Float)
    close = Column(Float)
    volume = Column(Integer)
    market_cap = Column(Float)
    unit_volume = Column(Float)
    rolling_20_d = Column(Float)
    date_2 = Column(String)
    bitfinex_shorts = Column(Float)
    bitfinex_longs = Column(Float)
    bitfinex_volume = Column(Float)
    total_crypto_cap = Column(Float)
    bitcoin_dominance = Column(Float)
```

```python
# Table only exists in python and not in the actual database yet
Base.metadata.tables
```

```python
engine = create_engine("sqlite:///./db/data.sqlite")
```

```python
# This is where the tables was created in the database
Base.metadata.create_all(engine)
```

```python
# The ORM's "handle" to the database is the Session
session = Session(engine)
```

```python
# Adding the data to the session

for i in btc_df.index:
    session.add(btc(
        date = btc_df["date"][i],
        open = btc_df["open"][i],
        high = btc_df["high"][i],
        low = btc_df["low"][i],
        close = btc_df["close"][i],
        volume = btc_df["volume"][i],
        market_cap = btc_df["market_cap"][i],
        unit_volume = btc_df["unit_volume"][i],
        rolling_20_d = btc_df["rolling_20_d"][i],
        date_2 = btc_df["date_2"][i],
        bitfinex_shorts = btc_df["bitfinex_shorts"][i],
        bitfinex_longs = btc_df["bitfinex_longs"][i],
        bitfinex_volume = btc_df["bitfinex_volume"][i],
        total_crypto_cap = btc_df["total_crypto_cap"][i],
        bitcoin_dominance = btc_df["bitcoin_dominance"][i]
        ))
```

```python
# Data has not been added to the database yet
engine.execute("SELECT * FROM btc").fetchall()
```

```python
# Use the new attribute to see the queue of data ready to go into the database
session.new
```

```python
# Commit flushes whatever remaining changes remain to the database and commits the transaction
session.commit()
```

```python
# Query the database to check the data
session.query(btc.rolling_20_d).all()
```
