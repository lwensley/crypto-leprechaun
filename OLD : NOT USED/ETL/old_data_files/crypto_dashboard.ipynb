```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
```

```python
"""html_string = 
<html>
<head>
<title>
A Simple HTML Document
</title>
</head>
<body>
<p>This is a very simple HTML document</p>
<p>It only has two paragraphs</p>
</body>
</html>
"""
url="https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20180505&end=20190505"
```

```python
# Retrieve page with the requests module
response = requests.get(url)  #response shd be 200
print(response)

# Create a Beautiful Soup object with html parses
soup = bs(response.text, 'html.parser') # beautiful soup, bs is an alias so we don't have to type it 

#html parser just knows it will by <> stuff upcoming 
type(soup)
# Print formatted version of the soup
#print(soup.prettify())
```

```python
# Extract the title of the HTML document
soup.title
```

```python
# Extract the text of the title
soup.title.text
```

```python
# Clean up the text
soup.title.text.strip()
```

```python
# Extract the contents of the HTML body
soup.body
```

```python
# Extract the text of the body
soup.body.text
soup.body.text.strip() # clean it up with the strip 
```

```python
# Text of the first paragraph - here p stands for paragraph
soup.body.p.text
```

```python
# Extract all paragraph elements
soup.body.find_all('table')
```

```python
# Extract paragraph by index
soup.body.find_all('p')[0]
```

```python
soup.body.find_all('p')[1] #this searches for the 2nd paragrah
```
